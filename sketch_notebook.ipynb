```python
import os
import pandas as pd

def concat_csv_files(folder_path):
    """
    Concatenate all CSV files in a folder into a single DataFrame.
    
    Parameters:
        folder_path (str): Path to the folder containing CSV files.
    
    Returns:
        pd.DataFrame: Concatenated DataFrame.
    """
    all_dataframes = []

    # Loop through files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            filepath = os.path.join(folder_path, filename)
            # Read CSV into a DataFrame, ignoring the index
            df = pd.read_csv(filepath, index_col=None)
            all_dataframes.append(df)

    # Concatenate all DataFrames into a single DataFrame
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    return combined_df


folder_path = "./data_gold/"
combined_dataframe = concat_csv_files(folder_path)
combined_dataframe.to_csv('df_gold.csv')
combined_dataframe
```
